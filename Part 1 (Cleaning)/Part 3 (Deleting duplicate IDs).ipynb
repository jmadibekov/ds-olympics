# Deleting duplicate IDs jupyter

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
data = pd.read_excel('Cleaned_Data_With_Duplicates_ID.xlsx')
data.head(10)
```

### Removing duplicates  (year,season,sport type,podium is the same)

```python
dup = {}

data2 = data.copy()
j = 0
for rows in data2.iterrows():
    key = (rows[1][0],rows[1][8],rows[1][11],rows[1][18],rows[1][9])
    if key in dup:
        data2.drop(j,inplace = True)
    else:
        dup[key]=1
    j += 1
```

### Exporting the file as without Duplicates

```python
data2.to_excel('Cleaned_Data_Without_Duplicates_ID.xlsx', index = None, header=True)
```
